```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load your dataset

```python
salaries = pd.read_csv('/content/Salaries.csv')
salaries.head()
```

# Task 1: Basic Data Exploration

```python
rows, columns = salaries.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")

data_types = salaries.dtypes
print("\nData Types:")
print(data_types)

missing_values = salaries.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

# Task 2: Descriptive Statistics

```python
basic_stats = salaries[['TotalPay']].describe()
salary_range = salaries['TotalPay'].max() - salaries['TotalPay'].min()
std_dev = salaries['TotalPay'].std()

print("\nBasic Statistics:")
print(basic_stats)
print("\nSalary Range:", salary_range)
print("\nStandard Deviation:", std_dev)
```

# Task 3: Data Cleaning

```python
median_salary = salaries['TotalPay'].median()
salaries['TotalPay'].fillna(median_salary, inplace=True)
```

# Task 4: Basic Data Visualization

```python
plt.hist(salaries['TotalPay'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()

department_counts = salaries['JobTitle'].value_counts().head(10)
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 10 Job Titles Distribution')
plt.show()
```

# Task 5: Grouped Analysis

```python
average_salaries_by_year = salaries.groupby('Year')['TotalPay'].mean()
print("\nAverage Salaries by Year:")
print(average_salaries_by_year)
```

# Task 6: Simple Correlation Analysis

```python
correlation = salaries['TotalPay'].corr(salaries['BasePay'])
print("\nCorrelation between TotalPay and BasePay:", correlation)

plt.scatter(salaries['BasePay'], salaries['TotalPay'], alpha=0.5)
plt.title('Scatter Plot: TotalPay vs BasePay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.show()
```
